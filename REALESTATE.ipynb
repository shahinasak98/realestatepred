```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:\\Bengaluru_House_Data.csv")
data.head(10)
```

```python
data.describe()
data1=data.drop(['society','balcony','availability','area_type'],axis='columns')
data1.head(30)
```

```python
data1.isnull().sum()
data2=data1.dropna()
data2
```

```python
func=lambda x : (int(x.split(' ')[0]))
data2['BHK']=data2['size'].apply(func)
data2.head(10)
```

```python
data3=data2.drop(['size'],axis='columns')
data3.head(5)
```

```python
data3['BHK'].unique()
data3[data3.BHK>10]
```

```python
data3['total_sqft'].unique()
```

```python
def isFloat(x):
    try:
        float(x)
    except:
        return False
    return True

data4=data3[~data3['total_sqft'].apply(isFloat)]
data4
# data4['total_sqft'].unique()
```

```python
import re
```

```python
def conversion(x):
    y=re.search("-",str(x))
    if(y):
        variables=x.split('-')
        return (float(variables[0])+float(variables[1]))/2
    else:
        return x
        
```

```python
data3['total_sqft']=data3['total_sqft'].apply(conversion)
data3.loc[30]
```

```python
# line = re.sub(r"[^a-zA-Z\s']")
# print(line)
data4=data3[~data3['total_sqft'].str.contains("[a-zA-Z]").fillna(False)]

```

```python
data5=data4.copy()
data5['total_sqft'].str.strip()
```

```python
data5['price_per_sqft']=(data5['price']*100000)
```

```python
data5['total_sqft']=pd.to_numeric(data5['total_sqft'],downcast="float")
```

```python
data5['price_per_sqft']=data5['price_per_sqft']/data5['total_sqft']
```

```python
len(data5['location'].unique())
```

```python
data5['location']=data5['location'].apply(lambda x: x.strip())
```

```python
location_count=data5.groupby('location')['location'].agg('count')
location_count
```

```python
len(location_count[location_count<=10])
```

```python
otherlist=location_count[location_count<=10]
```

```python
data5['location']=data5['location'].apply((lambda x: 'Other' if x in otherlist else x ))
```

```python
data6=data5.copy()
data6
```

```python
data7=data6[~((data6['total_sqft']/data5['BHK'])<300)]
data7
```

```python
data7['price_per_sqft'].describe()
```

```python
# Removing the outliers

def outliersremoval(dataframe):
    data8=pd.DataFrame()
    for counter,idf in dataframe.groupby('location'):
        mu=np.mean(idf['price_per_sqft'])
        sigma=np.std(idf['price_per_sqft'])
        outputdf=idf[(idf['price_per_sqft']>(mu-sigma)) & (idf['price_per_sqft']<=(mu+sigma))]
        data8=pd.concat([data8,outputdf],ignore_index=True)
    return data8
```

```python
data9=outliersremoval(data7)
data9
```

```python
data10=data9[~(data9['bath']>(data9['BHK'])+2)]
data10
```

```python
import matplotlib as ml
def plot_scatter_chart(df,location):
    bhk2 = df[(df['location']==location) & (df['BHK']==2)]
    bhk3 = df[(df['location']==location) & (df['BHK']==3)]
    ml.pyplot.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='red',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
    
plot_scatter_chart(data10,"Whitefield")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
data11 = remove_bhk_outliers(data10)
```

```python
plot_scatter_chart(data11,"Whitefield")
```

```python
#Hot encoding

dummies=pd.get_dummies(data11['location'])
dummies.head()
```

```python
dummiesf=dummies.drop('Other',axis='columns')
```

```python
data12=pd.concat([data11,dummiesf],axis='columns')
data12.head(15)
```

```python
data12=data12.drop('location',axis='columns')
```

```python
data12=data12.drop('price_per_sqft',axis='columns')
```

```python
#Model-Building
finaldataframe=data12.copy()
finaldataframe.describe()
```

```python
#Split in to X and Y variables
X=finaldataframe.drop('price',axis='columns')
y=finaldataframe['price']
X
```

```python
y
```

```python
# Split data into train vs test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
X_train=np.array(X_train.values)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
print(lr.score(X_test,y_test))
```

```python
from sklearn.model_selection import ShuffleSplit, cross_val_score
cv = ShuffleSplit(n_splits=5, test_size = 0.2, random_state=0)
cross_val_score(lr,X,y,cv=cv)
```

```python
#To check accuracy of the model
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
cross_val_score(LinearRegression(),X,y,cv=cv)

# Scores are above 80% - Which is good for the model.
```

```python
# we can use Gridsearhcv to determine how accurate is our project

from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearch(X,y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': { 'normalize': [True, False]}
            },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random','cyclic']
                }
            },
        'decision_tree':{
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse','friedman_mse'],
                'splitter': ['best','random']
                }
            }
        }
    scores = []
    cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
    for algo_name,config in algos.items():
        gs = GridSearchCV(config['model'],
                          config['params'],
                          cv=cv,
                          n_jobs=-1,
                          return_train_score=False
                          )
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
            })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
model_scores = find_best_model_using_gridsearch(X,y)
print(model_scores)
```

```python
from sklearn.linear_model import LinearRegression
regress = LinearRegression()
regress.fit(X,y)
```

```python
# We can predict prices now
def predict_price(location,sqft,bath,bhk):
    loc_index = np.where(X.columns == location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    # print(loc_index,"locindex")
    if loc_index >= 0:
        x[loc_index] = 1
    return round((regress.predict([x])[0]),2)

   
```

```python
print(predict_price('1st Phase JP Nagar',1000,2,2))
print(predict_price('1st Phase JP Nagar',1000,3,3))
print(predict_price('Indira Nagar',2500,4,4))
```
